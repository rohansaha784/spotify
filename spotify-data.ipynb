```python
# Fetch the user's top tracks

top_tracks = sp.current_user_top_tracks(time_range='medium_term', limit=50)
top_tracks_1 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 50)
top_tracks_2 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 100)
top_tracks_3 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 150)
top_tracks_4 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 200)
top_tracks_5 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 250)
top_tracks_6 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 300)
top_tracks_7 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 350)
top_tracks_8 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 400)
top_tracks_9 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 450)
top_tracks_10 = sp.current_user_top_tracks(time_range='medium_term', limit=50, offset = 500)

lists = [top_tracks, top_tracks_1, top_tracks_2, top_tracks_3, top_tracks_4, top_tracks_5,
         top_tracks_6, top_tracks_7, top_tracks_8, top_tracks_9, top_tracks_10]

songs = []

# Print each track's name and ID
for top_tracks in lists:
    for track in top_tracks['items']:
        print(track['name'], '-', track['id'])
        artist_list = []
        for item in track['artists']:
            artist_list.append(item['name'])
        artist_list = ', '.join(artist_list)
        songs.append([track['name'], track['id'], artist_list])
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=cli_id,
    client_secret=cli_secret,
    redirect_uri=url,
    scope='user-read-private'))



for song in songs:
    
    song_id = song[1]
    audio_features = sp.audio_features(song_id)[0]
    is_explicit = int(sp.track(song_id)['explicit'])
    
    if len(song) < 15:
        song.append(audio_features['danceability'])
        song.append(audio_features['energy'])
        song.append(audio_features['key'])
        song.append(audio_features['loudness'])
        song.append(audio_features['mode'])
        song.append(audio_features['speechiness'])
        song.append(audio_features['acousticness'])
        song.append(audio_features['instrumentalness'])
        song.append(audio_features['liveness'])
        song.append(audio_features['valence'])
        song.append(audio_features['tempo'])
        song.append(audio_features['time_signature'])
        song.append(audio_features['duration_ms'])
        song.append(is_explicit)
        print(song)

```

```python
song_info = pd.DataFrame(songs, columns = ['name', 'id', 'artists','danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 
                                           'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'duration_ms', 'is_explicit'])
song_info = song_info.reset_index()
song_info['ranking'] = song_info['index'] + 1
song_info = song_info.drop(['index', 'id'], axis = 1)
song_info.head(10)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler1 = StandardScaler()
scaler2 = MinMaxScaler()
df_filtered = song_training_data[['energy', 'loudness', 'danceability', 'duration_ms', 'valence','instrumentalness', 'tempo', 'ranking']]
df_minmax = pd.DataFrame(scaler2.fit_transform(df_filtered[['energy', 'ranking', 'danceability', 'valence','instrumentalness']]), columns = ['energy', 'ranking', 'danceability', 'valence','instrumentalness'])
df_standard = pd.DataFrame(scaler1.fit_transform(df_filtered[['loudness', 'duration_ms', 'tempo']]), columns = ['loudness', 'duration_ms', 'tempo'])
df_filtered = pd.concat([song_info[['name', 'artists']], df_minmax, df_standard], axis = 1)
df_filtered.info()
```

```python
import musicbrainzngs
import numpy as np

from spotipy.oauth2 import SpotifyClientCredentials

musicbrainzngs.set_useragent("Trendify.ai", "1.0", "harshilmachhi18@gmail.com")

def get_artist_country(artist_name):
    """Fetch the country of origin for an artist from MusicBrainz."""
    try:
        
        result = musicbrainzngs.search_artists(artist=artist_name, limit=1)

        if result['artist-list']:
            artist = result['artist-list'][0]
            country = artist.get('country', 'Unknown')
            if country == "Unknown":
                return 'US'
            return country
        else:
            return 'US'
    except Exception as e:
        return 'US'


df_filtered['country'] = df_filtered['artists'].apply(lambda x: get_artist_country(x.split(", ")[0]))
```
